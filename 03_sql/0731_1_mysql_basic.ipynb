* 테이블 생성
  - Cursor Object 가져오기: cursor = db.cursor()  
  - SQL 실행하기: cursor.execute(SQL)
  - 실행 mysql 서버에 확정 반영하기: db.commit()

````python
import pymysql
connection = pymysql.connect(host="localhost", user="root", password="", db="fisa")

SQL = """
CREATE TABLE IF NOT EXISTS box_office_two (
    movie_code VARCHAR(20),
    movie_name VARCHAR(100),
    release_date DATE,
    sale_amt INT
)
"""

connection.cursor().execute(SQL)

connection.close()
````

* cursor 는 control structure of database 입니다. (연결된 객체로 보셔도 좋습니다.)

````python
import pymysql
from pymysql.constants import CLIENT
import pandas as pd

connection = pymysql.connect(
  host="localhost",
  user="root",
  password="",
  db="student_mgmt",
  charset='utf8',
  client_flag=CLIENT.MULTI_STATEMENTS
)

SQL = """
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('wage2', 'woman', '1982-1-13', 76, 30, 80);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('tina2', 'woman', '1982-12-3', 87, 62, 71)
"""

connection.cursor().execute(SQL)
connection.commit()

SQL2 = "SELECT * from students"
read = pd.read_sql(SQL2, connection)

connection.close()
read
````

````python
# update
import pymysql
import pandas as pd

connection = pymysql.connect(
    host="localhost", user="root", password="", db="student_mgmt"
)

SQL = """
UPDATE students SET english = 100 WHERE name = '김철수'
"""

connection.cursor().execute(SQL)
connection.commit()

SQL2 = "SELECT * from students"
read = pd.read_sql(SQL2, connection)

connection.close()
read
````

````python
# delete
import pymysql
import pandas as pd

connection = pymysql.connect(
    host="localhost", user="root", password="", db="student_mgmt"
)

SQL = """
DELETE FROM students WHERE name = '김철수'
"""

connection.cursor().execute(SQL)
connection.commit()

SQL2 = "SELECT * from students"
read = pd.read_sql(SQL2, connection)

connection.close()
read
````

* 삽입, 갱신, 삭제 등이 모두 끝났으면 Connection 객체의 commit() 메서드를 사용하여 데이타를 Commit

* Connection 객체의 close() 메서드를 사용하여 DB 연결을 닫음

````python
import pymysql
from pymysql.constants import CLIENT
import pandas as pd

db = pymysql.connect(
    host="localhost",  # 127.0.0.1 -> 자기자신을 가리킴
    port=3306,
    user="root",
    passwd="",
    db="student_mgmt",
    charset="utf8",
    client_flag=CLIENT.MULTI_STATEMENTS,
    )

student_mgmt = db.cursor()
sql3 = """
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('davey', 'man', '1983-07-16', 90, 80, 71);
INSERT INTO students (name, gender, birth, english, math, korean) VALUES ('tina2', 'woman', '1982-12-3', 87, 62, 71);
"""

# # 5. SQL 구문 실행하기
student_mgmt.execute(sql3)

# 6. DB에 Complete 하기
db.commit()
# 7. DB 연결 닫기
db.close()
````

### 패턴으로 익히는 pymysql

* 데이터 조회(SELECT)
  - Cursor Object 가져오기: cursor = db.cursor()  
  - SQL 실행하기: cursor.execute(SQL)
  - mysql 서버로부터 데이터 가져오기: fetch 메서드 사용
    - fetchall(): Fetch all the rows
    - fetchmany(size=None): Fetch several rows
    - fetchone(): Fetch the next row

````python
import pymysql
from pymysql.constants import CLIENT
import pandas as pd

db = pymysql.connect(
    host="localhost",  # 127.0.0.1 -> 자기자신을 가리킴
    port=3306,
    user="root",
    passwd="",
    db="student_mgmt",
    charset="utf8",
    client_flag=CLIENT.MULTI_STATEMENTS,
)

# Using the fetchall() method:
cursor = db.cursor()
cursor.execute("SELECT * FROM students")
fetchall = cursor.fetchall()

# Using the fetchone() method:
cursor.execute("SELECT * FROM students")
fetchone = cursor.fetchone()

# Using the fetchmany() method:
cursor.execute("SELECT * FROM students")
fetchmany = cursor.fetchmany(size=5)

db.close()

# Print the result by pandas
# pd.DataFrame(fetchall)
# pd.DataFrame(fetchone)
pd.DataFrame(fetchmany)
````

* 데이터 수정(UPDATE)
  - Cursor Object 가져오기: cursor = db.cursor()  
  - SQL 실행하기: cursor.execute(SQL)
  - 실행 mysql 서버에 확정 반영하기: db.commit()

* 데이터 삭제(DELETE)
  - Cursor Object 가져오기: cursor = db.cursor()  
  - SQL 실행하기: cursor.execute(SQL)
  - 실행 mysql 서버에 확정 반영하기: db.commit()
